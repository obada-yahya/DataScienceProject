```python
import pandas as pd 
import matplotlib.pyplot as plot
```

```python
# TODO.1 load it into a pandas dataframe using the read_csv() function
data = pd.read_csv("./imdb_top_1000.csv")
```

```python
data.head()
```

```python
# TODO.2 Clean and preprocess the data by removing unnecessary columns and handling missing values.
data = data[["Director", "Series_Title", "Genre"]]
```

```python
# TODO.3 Most Popular genre of movies and TV Shows
print(data["Genre"].value_counts().index[0])
```

```python
# TODO.4 Create a new dataframe containing the count of movies/TV shows in each genre using pandas.
GenreCounts = data["Genre"].value_counts()
newDataFrame = pd.DataFrame({'Genre': GenreCounts.index, 'Total': GenreCounts.tolist()}, index=range(GenreCounts.shape[0]))
print(newDataFrame)
```

```python
# TODO.5 Visualize the data using a bar chart to show the most popular genre of movies/TV
# we have 202 records , it's impossible to view them , so i selected the first 10 records
newDataFrame[:10].plot.bar(x="Genre", y="Total", title="Most popular genre of movies/TV")
plot.show(block=True)
```

```python
# TODO.6 Identify the director with the most top-rated movies/TV shows by creating a new
topDirectors = data["Director"].value_counts()
print(topDirectors.index[0], topDirectors[0])
topDirectorsDataFrame = pd.DataFrame({"Director": topDirectors.index, 'Total': topDirectors.tolist()}, index=range(topDirectors.shape[0]))
```

```python
# TODO.7 Visualize the data using a bar chart to show the director with the most top-rated movies/TV
# we have 548 records , it's impossible to view them , so i selected the first 10 records
topDirectorsDataFrame[:10].plot.bar(x="Director", y="Total", title="Most Popular Directors of movies/TV")
plot.show(block=True)
```
